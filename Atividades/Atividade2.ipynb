Data Dictionary  
`Booking_ID`: unique identifier of each booking  
`no_of_adults`: Number of adults  
`no_of_children`: Number of Children  
`no_of_weekend_nights`: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel  
`no_of_week_nights`: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel  
`type_of_meal_plan`: Type of meal plan booked by the customer:  
`required_car_parking_space`: Does the customer require a car parking space? (0 - No, 1- Yes)  
`room_type_reserved`: Type of room reserved by the customer. The values are ciphered (encoded) by INN Hotels.  
`lead_time`: Number of days between the date of booking and the arrival date  
`arrival_year`: Year of arrival date  
`arrival_month`: Month of arrival date  
`arrival_date`: Date of the month  
`market_segment_type`: Market segment designation.  
`repeated_guest`: Is the customer a repeated guest? (0 - No, 1- Yes)  
`no_of_previous_cancellations`: Number of previous bookings that were canceled by the customer prior to the current booking  
`no_of_previous_bookings_not_canceled`: Number of previous bookings not canceled by the customer prior to the current booking  
`avg_price_per_room`: Average price per day of the reservation; prices of the rooms are dynamic. (in euros)  
`no_of_special_requests`: Total number of special requests made by the customer (e.g. high floor, view from the room, etc)  
`booking_status`: Flag indicating if the booking was canceled or not.

### (4) Desenvolver em Python um pipeline que faça as combinações de todos os classificadores com todas as técnicas de seleção de carcatéricas.
* Considere os seguintes classificadores: KNN, Ávore de Decisão e Randon Forest (biblioteca Sklearn)
* Considere as seguintes técnicas de seleção de atributos: Mutual Information e Anova (Sklearn) e SmartCorrelatedGroups (Feature Engine)  
Ao final, apresentar o Diagrama de Venn com os conjuntos das 3 melhores combinações.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
from feature_engine.selection import SmartCorrelatedSelection
```

```python
X = df_scaled.drop(['booking_status'], axis=1)
y = df_scaled['booking_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

#### Configuração das técnicas que serão utilizadas

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree_pred = tree.predict(X_test)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)

forest = RandomForestClassifier(n_estimators = 100)
forest.fit(X_train, y_train)
forest_pred = forest.predict(X_test)


mi = SelectKBest(mutual_info_classif, k=15)

anova = SelectKBest(f_classif, k=15)

smartC = SmartCorrelatedSelection(
    method="pearson",
    threshold=0.8,
    selection_method="variance",
    estimator=None
)
```

```python
ppl1 = Pipeline([('mi_s', mi), ('knn_c', knn)])
ppl1.fit(X_train, y_train)
ppl1_pred = ppl1.predict(X_test)

ppl2 = Pipeline([('mi_s', mi), ('dt_c', tree)])
ppl2.fit(X_train, y_train)
ppl2_pred = ppl2.predict(X_test)

ppl3 = Pipeline([('mi_s', mi), ('rf_c', forest)])
ppl3.fit(X_train, y_train)
ppl3_pred = ppl3.predict(X_test)

ppl4 = Pipeline([('anova_s', anova), ('knn_c', knn)])
ppl4.fit(X_train, y_train)
ppl4_pred = ppl4.predict(X_test)

ppl5 = Pipeline([('anova_s', anova), ('dt_c', tree)])
ppl5.fit(X_train, y_train)
ppl5_pred = ppl5.predict(X_test)

ppl6 = Pipeline([('anova_s', anova), ('rf_c', forest)])
ppl6.fit(X_train, y_train)
ppl6_pred = ppl6.predict(X_test)

ppl7 = Pipeline([('smartc_s', smartC), ('knn_c', knn)])
ppl7.fit(X_train, y_train)
ppl7_pred = ppl7.predict(X_test)

ppl8 = Pipeline([('smartc_s', smartC), ('dt_c', tree)])
ppl8.fit(X_train, y_train)
ppl8_pred = ppl8.predict(X_test)

ppl9 = Pipeline([('smartc_s', smartC), ('rf_c', forest)])
ppl9.fit(X_train, y_train)
ppl9_pred = ppl9.predict(X_test)
```

```python
dct_pred = {
    'PPL1 - MI+KNN': ppl1_pred,
    'PPL2 - MI+DT': ppl2_pred,
    'PPL3 - MI+RF': ppl3_pred,
    'PPL4 - Anova+KNN': ppl4_pred,
    'PPL5 - Anova+DT': ppl5_pred,
    'PPL6 - Anova+RF': ppl6_pred,
    'PPL7 - SmartCorr+KNN': ppl7_pred,
    'PPL8 - SmartCorr+DT': ppl8_pred,
    'PPL9 - SmartCorr+RF': ppl9_pred
}


for pipe in dct_pred:
    # print(pipe)
    cm = confusion_matrix(y_test, dct_pred[pipe])
    ConfusionMatrixDisplay(cm).plot()
    plt.title(pipe)
    plt.show()
    
```

```python
print(f"Pipelines com KNN")
print(f"{' ':^10s} | {'ACC':^10s} | {'Recall':^10s} | {'Precision':^10s} | {'F1-Score':^10s}")
print(f"{'Completo':<10s} | {accuracy_score(y_test,knn_pred):10.4f} | {recall_score(y_test,knn_pred):10.4f} | {precision_score(y_test,knn_pred):10.4f} | {f1_score(y_test,knn_pred):10.4f}")
print(f"{'MI':<10s} | {accuracy_score(y_test,ppl1_pred):10.4f} | {recall_score(y_test,ppl1_pred):10.4f} | {precision_score(y_test,ppl1_pred):10.4f} | {f1_score(y_test,ppl1_pred):10.4f}")
print(f"{'Anova':<10s} | {accuracy_score(y_test,ppl4_pred):10.4f} | {recall_score(y_test,ppl4_pred):10.4f} | {precision_score(y_test,ppl4_pred):10.4f} | {f1_score(y_test,ppl4_pred):10.4f}")
print(f"{'SmartC':<10s} | {accuracy_score(y_test,ppl7_pred):10.4f} | {recall_score(y_test,ppl7_pred):10.4f} | {precision_score(y_test,ppl7_pred):10.4f} | {f1_score(y_test,ppl7_pred):10.4f}")
print('-=-'*21)
print(f"Pipelines com Decision Tree")
print(f"{' ':^10s} | {'ACC':^10s} | {'Recall':^10s} | {'Precision':^10s} | {'F1-Score':^10s}")
print(f"{'Completo':<10s} | {accuracy_score(y_test,tree_pred):10.4f} | {recall_score(y_test,tree_pred):10.4f} | {precision_score(y_test,tree_pred):10.4f} | {f1_score(y_test,tree_pred):10.4f}")
print(f"{'MI':<10s} | {accuracy_score(y_test,ppl2_pred):10.4f} | {recall_score(y_test,ppl2_pred):10.4f} | {precision_score(y_test,ppl2_pred):10.4f} | {f1_score(y_test,ppl2_pred):10.4f}")
print(f"{'Anova':<10s} | {accuracy_score(y_test,ppl5_pred):10.4f} | {recall_score(y_test,ppl5_pred):10.4f} | {precision_score(y_test,ppl5_pred):10.4f} | {f1_score(y_test,ppl5_pred):10.4f}")
print(f"{'SmartC':<10s} | {accuracy_score(y_test,ppl8_pred):10.4f} | {recall_score(y_test,ppl8_pred):10.4f} | {precision_score(y_test,ppl8_pred):10.4f} | {f1_score(y_test,ppl8_pred):10.4f}")
print('-=-'*21)
print(f"Pipelines com Random Forest")
print(f"{' ':^10s} | {'ACC':^10s} | {'Recall':^10s} | {'Precision':^10s} | {'F1-Score':^10s}")
print(f"{'Completo':<10s} | {accuracy_score(y_test,forest_pred):10.4f} | {recall_score(y_test,forest_pred):10.4f} | {precision_score(y_test,forest_pred):10.4f} | {f1_score(y_test,forest_pred):10.4f}")
print(f"{'MI':<10s} | {accuracy_score(y_test,ppl3_pred):10.4f} | {recall_score(y_test,ppl3_pred):10.4f} | {precision_score(y_test,ppl3_pred):10.4f} | {f1_score(y_test,ppl3_pred):10.4f}")
print(f"{'Anova':<10s} | {accuracy_score(y_test,ppl6_pred):10.4f} | {recall_score(y_test,ppl6_pred):10.4f} | {precision_score(y_test,ppl6_pred):10.4f} | {f1_score(y_test,ppl6_pred):10.4f}")
print(f"{'SmartC':<10s} | {accuracy_score(y_test,ppl9_pred):10.4f} | {recall_score(y_test,ppl9_pred):10.4f} | {precision_score(y_test,ppl9_pred):10.4f} | {f1_score(y_test,ppl9_pred):10.4f}")
print('-=-'*21)
```

Podemos portanto perceber que em todas as métricas e em todas os algoritmos de seleção de algotirmos, o classificador `Random Forest` tem um desempenho superior.
